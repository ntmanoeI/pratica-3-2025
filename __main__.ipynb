**importe a biblioteca pandas**

```python
import pandas as pd
```

**crie uma variável ("data") e leia o conteúdo do arquivo CSV, passando como parâmetros o separador de colunas (;), a engine - com o valor ‘python’ e o enconding relativo aos dados constantes no arquivo lido (esse último parâmetro pode ser opcional, dependendo do enconding existente); atribua os dados lidos do CSV à variável criada anteriormente; salve as alterações**

```python
data = pd.read_csv('picoweb.csv', sep=';', engine='python', encoding='utf-8')
```

**imprimir as informações gerais sobre o conjunto de dados.**

```python
print("Informações gerais sobre o conjunto de dados:")
display(data.info())
```

**imprimir as 10 primeiras linhas do arquivo**

```python
print("Imprimir as primeiras 10 linhas:")
display(data.head(10))
```

**imprimir as 10 últimas linhas do arquivo**

```python
print("Imprimir as últimas 10 linhas:")
display(data.tail(10))
```

**criar uma nova variável (data_copy) e atribuir a ela uma cópia do conjunto de dados original**

```python
data_copy = data.copy()
```

**substituir todos os valores nulos da coluna ‘Calories’ por 0**

```python
data_copy['Calories'] = data_copy['Calories'].fillna(0)
print("Dados após substituir nulos em 'Calories':")
display(data_copy)
```

**substituir os valores nulos da coluna 'Date' por '1900/01/01'.**

```python
data_copy['Date'] = data_copy['Date'].fillna('1900/01/01')
print("Dados após substituir nulos em 'Date':")
display(data_copy)
```

**transformar os dados da coluna 'Date' em datetime usando o método 'to_datetime'**

```python
data_copy['Date'] = pd.to_datetime(data_copy['Date'], errors='coerce')
print("Dados após a conversão inicial para datetime:")
display(data_copy)
```

**substituir, na coluna 'Date', o valor '1900/01/01' por 'NaT'**

```python
data_copy['Date'] = data_copy['Date'].replace('1900-01-01', pd.NaT)
print("Dados após substituir '1900/01/01' por NaT:")
display(data_copy)
```

**utilizando o método 'to_datetime', repita o passo de transformação  dos dados da coluna 'Date' para datetime**

```python
data_copy['Date'] = pd.to_datetime(data_copy['Date'], errors='coerce')
print("Dados após a nova conversão para datetime:")
display(data_copy)
```

**combinar os métodos 'replace' e 'to_datetime'**

```python
data_copy['Date'] = data_copy['Date'].replace('20201226', '2020/12/26')
data_copy['Date'] = pd.to_datetime(data_copy['Date'], errors='coerce')
print("Dados após corrigir '20201226' e converter para datetime:")
display(data_copy)
```

**remover os registros contendo valores nulos**

```python
data_copy.dropna(subset=['Date'], inplace=True)
```

```python
print("Dados finais após todas as transformações e remoção de nulos:")
display(data_copy)
```
